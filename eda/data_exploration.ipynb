```python
import configparser
import pandas as pd
import io

def init_config():
    config = configparser.ConfigParser()
    config.read("../config.ini")
    return config

def download_blob(config):
    """Downloads a blob from the bucket."""
    
    # Initialize a storage client
    path = config['gcp']['cloud_storage']
    df = pd.read_csv(path)
    return df
```

```python
config = init_config()
file = download_blob(config)
```
